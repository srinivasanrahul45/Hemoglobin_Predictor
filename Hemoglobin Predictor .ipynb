## Import Dataset and basic preprocessing



```python
dataset = pd.read_csv('/content/Dataset - Final.csv')
dataset.head(10)
```

```python
#Changing pandas dataframe to numpy array
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

## Feature Scaling and Splitting the dataset

```python
# Splitting dataset
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.1)
```

```python
print(f"X-Train is {X_train.shape}")
print(f"X-Test is {X_test.shape}")
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
sc = MinMaxScaler(feature_range = (0,1))
# sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

## Building the Neural Network

```python
# Building ANN
#Dependencies
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Neural network
# sigmoid, relu
model = Sequential()
model.add(Dense(20,activation='relu',input_dim=5))
model.add(Dense(20,activation='relu'))
model.add(Dense(20,activation='relu'))
model.add(Dense(1))
```

```python
model.compile(loss='mean_squared_error', optimizer='adam')
```

## Training the model

```python
history = model.fit(X_train, y_train, validation_data = (X_test,y_test), epochs=500)
```

```python
print(history.history.values())
```

## Testting the model

### Single value predection

```python
t = sc.fit_transform([[3,2,20.21,41.3,27]])
y_pred = model.predict(t)
#  y_pred = sc.inverse_transform(y_pred)
print(y_pred)
```

## Evaluating model performance (Debugging)

### Measuring Accuracy

```python
accuracy = model.evaluate(X_test, y_test)
print('Accuracy: %.2f' % ((1-accuracy)*100))
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss']) 
plt.title('Model loss') 
plt.ylabel('Loss') 
plt.xlabel('Epoch') 
plt.legend(['Train', 'Validation'], loc='upper left') 
plt.show()
```
